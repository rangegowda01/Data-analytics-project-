## Step 6 : Data Visualization

```python
# to creates a DataFrame containing the first 20 rows of the original data.
df = pd.DataFrame(data.head(20))

# Calculate the count of reviews for each star rating
reviews_by_rating = df['stars'].value_counts()

# Create a pie chart to display the distribution of reviews by star rating
reviews_by_rating.plot(kind='pie', title='Reviews by Star Rating', autopct='%1.1f%%')

# Display the pie chart
plt.show()
```

```python
# Create a DataFrame containing the first 15 rows of the original data.
df = pd.DataFrame(data.head(15))

# Calculate the average rating for each brand
avg_rating_by_brand = df.groupby('brand')['stars'].mean()

# Create a bar chart to display the average rating by brand
avg_rating_by_brand.plot(kind='bar', title='Average Rating by Brand' , color='Orange')

# Add labels for the x and y axes
plt.xlabel('Brand')
plt.ylabel('Average Rating')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=55)

# Display the bar chart
plt.show()
```

```python
# Create a DataFrame containing the first 20 rows of the original data.
df = pd.DataFrame(data.head(20))

# Create a numerical index for the DataFrame
df.reset_index(drop=True, inplace=True)

# Plot the line chart

# Plotting price over index
plt.plot(df.index, df['price/value'], marker='o', color='Red')

# Setting the title of the plot
plt.title('Price Over Index')

# Labeling the x-axis
plt.xlabel('Index')

# Labeling the y-axis
plt.ylabel('Price')

# Displaying the plot
plt.show()
```

```python
# to creates a DataFrame containing the first 20 rows of the original data.
df = pd.DataFrame(data.head(15))

# calculates the frequency of each unique value in the 'stars' column of the DataFrame and display the percentage of each slice.
df['stars'].value_counts().plot(kind='hist', color='Yellow')

# Set the label for the x-axis
plt.xlabel('stars')

# Set the label for the y-axis
plt.ylabel('price/value')

# to give the title of the plot
plt.title('Frequency of Stars')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=0)

# to display the histogram chart
plt.show()  
```

```python
# Create a DataFrame containing the first 20 rows of the original data.
df = pd.DataFrame(data.head(20))

# Create a pivot table to aggregate reviews count by brand and star rating
reviews_pivot = df.pivot_table(index='brand', columns='stars', values='reviewsCount', aggfunc='sum', fill_value=0)

# Create a horizontal stacked bar chart to visualize reviews count by star rating and brand
reviews_pivot.plot(kind='barh', stacked=True)

# Set the title of the plot
plt.title('Reviews Count by Star Rating and Brand')

# Set the labels for the x-axis and y-axis
plt.xlabel('Brand')
plt.ylabel('Reviews Count')

# Display the horizontal stacked bar chart
plt.show()
```

```python
# Create a DataFrame containing the first 20 rows of the original data.
df = pd.DataFrame(data.head(20))

# Create a scatter plot
df.plot(kind='scatter', x='price/value', y='stars', title='Price vs. Stars', color='Brown')

# Label for the x-axis
plt.xlabel('Price')

# Label for the y-axis
plt.ylabel('Stars')

# Display the scatter plot
plt.show()
```
